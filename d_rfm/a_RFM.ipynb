```python
import numpy as np

Q1 = cp_df.describe().T.loc['Income', '25%']
Q3 = cp_df.describe().T.loc['Income', '75%']

iqr = Q3 - Q1

lower_bound = Q1 - iqr * 1.5
upper_bound = Q3 + iqr * 1.5

# ÏÜåÎìùÏù¥ ÏùåÏàòÏùº Ïàú ÏóÜÍ∏∞ÎïåÎ¨∏Ïóê ÏùåÏàòÏùº Í≤ΩÏö∞ 0ÏúºÎ°ú Î≥ÄÍ≤Ω
if lower_bound < 0 :
    lower_bound = 0

print(f'Ï†ïÏÉÅÏπò Î≤îÏúÑ: {lower_bound} ~ {upper_bound}')
```

#### Ï†ïÍ∑úÌôî (Nomalization)
Í∞íÏùò Î≤îÏúÑÎ•º 0~1 ÏÇ¨Ïù¥Î°ú Î≥ÄÌôòÏãúÏºúÏÑú Î™®Îì† Ïª¨ÎüºÏùò Îç∞Ïù¥ÌÑ∞Í∞Ä ÌèâÎì±ÌïòÍ≤å ÎêúÎã§.
ÏÑúÎ°ú Îã§Î•∏ Îã®ÏúÑÏùò Í∞íÏùÄ ÎπÑÍµê ÎåÄÏÉÅÏù¥ Îê† Ïàò ÏóÜÎã§. ÏòàÎ•º Îì§Ïñ¥, 80kgÍ≥º 180cmÎäî ÎπÑÍµêÌï† Ïàò ÏóÜÍ∏∞Ïóê Ï†ïÍ∑úÌôîÎ•º ÏÇ¨Ïö©Ìï¥ÏÑú ÎπÑÍµêÌïúÎã§.

```python
from sklearn.preprocessing import MinMaxScaler

# Î™®Îì† Í∞íÏùÑ 0~1ÏÇ¨Ïù¥Ïùò Í∞íÏúºÎ°ú Î≥ÄÍ≤Ω
nomalization = MinMaxScaler()
rfm_nomalization = nomalization.fit_transform(cp_rfm_df)
```

```python
cp_rfm_df = pd.DataFrame(rfm_nomalization, columns=cp_rfm_df.columns)
cp_rfm_df
```

```python
cp_df['TotalScore'] = cp_rfm_df['Recency'] + cp_rfm_df['Frequency'] + cp_rfm_df['Monetary']
cp_df
```

```python
# Ïù¥Ï†ú ÌïÑÏöîÏóÜÏñ¥ÏßÑ RFMÏùÑ DROP
cp_df = cp_df.drop(columns=['Recency', 'Frequency', 'Monetary'], axis=1)
cp_df
```

```python
import numpy as np

l1, l2, l3, l4 = np.percentile(cp_df['TotalScore'], [20, 40, 70, 90])
print(l1, l2, l3, l4)
```

```python
def get_level(x):
    if x <= l1:
        return 5
    if x > l1 and x <= l2:
        return 4
    if x > l2 and x <= l3:
        return 3
    if x > l3 and x <= l4:
        return 2
    return 1
```

```python
cp_df['Level'] = cp_df['TotalScore'].apply(get_level)
```
