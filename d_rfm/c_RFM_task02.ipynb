### RFM 분석 실습
- Recency: 얼마나 최근에 구매했는가
- Frequency: 얼마나 자주 구매했는가
- Monetary: 얼마나 많은 금액을 지출했는가

### 고객 분석
- InvoiceNo: 거래 번호
- StockCode: 상품 코드
- Description: 상품명 
- InvoiceDate: 거래 날짜
- UnitPrice: 상품 단가
- CustomerID: 고객 번호
- Country: 국가명

```python
import pandas as pd
import numpy as np
import datetime as dt

co_df = pd.read_csv('./datasets/customer_invoice_data.csv')
co_df.info()
```

```python
co_df['Quantity'].min()
```

```python
co_df
```

```python
# 오늘(2011-12-09)로부터 지난 구매 날짜 구하기
# datetype 타입으로 변경
co_df['InvoiceDate'] = pd.to_datetime(co_df['InvoiceDate'])
# 날짜 형식 변경
co_df['InvoiceDate'] = co_df['InvoiceDate'].dt.strftime('%Y-%m-%d')

co_df['Date'] = pd.to_datetime(co_df['InvoiceDate']).apply(lambda x: (pd.to_datetime('2011-12-09') - x).days)
co_df
```

```python
type(co_df['CustomerID'][0])
```

```python
# 결측치 제거
# 결측치가 많지만 CustomerID를 다른 값으로 대체하기엔 결과가 이상하게 나올 것같아 삭제
# 노트북 성능 이슈로 인해 경량하기 위함...
co_df = co_df[~ co_df['CustomerID'].isna()]
co_df.isna().sum()
```

```python
# co_df['CustomerID'] = co_df['CustomerID'].astype(np.int64)
# co_df['CustomerID'] = pd.DataFrame(co_df['CustomerID'])
# co_df
```

```python
# 0 이하의 값 제거
co_df = co_df[co_df['Quantity'] > 0]
```

```python
import numpy as np

Q1 = co_df.describe().T.loc['UnitPrice', '25%']
Q3 = co_df.describe().T.loc['UnitPrice', '75%']

iqr = Q3 - Q1

lower_bound = Q1 - iqr * 1.5
uppper_bound = Q3 + iqr * 1.5

if lower_bound < 0:
    lower_bound = 0

print(f'정상치 범위: {lower_bound} ~ {uppper_bound}')
```

```python
# unitPrice 이상치 제거
co_df = co_df[co_df['UnitPrice'] <= 7.5]
co_df
```

```python
co_df['StockCode'].value_counts()
```

```python
# Recency
rfm_r = co_df.groupby('CustomerID')['Date'].min().reset_index(name='Recency')
rfm_r
```

```python
# Frequency
rfm_f = co_df['CustomerID'].value_counts().reset_index(name='Frequency')
rfm_f
```

```python
# 구매수량과 가격을 곱하여 total_price를 만듬
total_price = co_df['Quantity'] * co_df['UnitPrice']
display(total_price)

co_df['TotalPrice'] = total_price

co_df
```

```python
# Monetary
rfm_m = co_df.groupby('CustomerID')['TotalPrice'].sum().reset_index(name='Monetary')
rfm_m
```

```python
# country = co_df.groupby('CustomerID')['Country'].apply(lambda x: ', '.join(x)).reset_index(name='Country')
# country
```

```python
# 테이블 병합
rfm = pd.merge(rfm_r, rfm_f, on='CustomerID')
rfm = pd.merge(rfm, rfm_m, on='CustomerID')
# rfm = pd.merge(rfm, country, on='CustomerID')
```

```python
rfm
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
rfm[['Recency', 'Frequency', 'Monetary']] = \
                                    scaler.fit_transform(rfm[['Recency', 'Frequency', 'Monetary']])
```

```python
rfm['Recency'] = 1 - rfm['Recency']
rfm
```

```python
# RFM 점수 total
rfm['TotalScore'] = rfm[['Recency', 'Frequency', 'Monetary']].sum(axis=1)
rfm
```

```python
# 기존 dataFrame에 merge
co_rfm_df = pd.merge(co_df, rfm, on='CustomerID')
co_rfm_df
```

```python
import numpy as np

# 위에서 구한 TotalScore를 기준으로 등급매기기
l1, l2, l3, l4 = np.percentile(co_rfm_df['TotalScore'], [20, 40, 77, 87])
print(l1, l2, l3, l4)
```

```python
def get_level(x):
    if x <= l1:
        return 5
    if x > l1 and x <= l2:
        return 4
    if x > l2 and x <= l3:
        return 3
    if x > l3 and x <= l4:
        return 2
    return 1
```

```python
co_rfm_df['Level'] = co_rfm_df['TotalScore'].apply(get_level)
```

```python
co_rfm_df
```

```python
co_rfm_df['Level'] = co_rfm_df['Level'].replace([5, 4, 3, 2, 1], ['Bronze', 'Silver', 'Gold', 'Diamond', 'VIP'])
```

```python
co_rfm_df
```

```python
print(co_rfm_df['Level'].value_counts())
```
