```python
import pandas as pd
import numpy as np
from scipy.stats import iqr

# https://www.kaggle.com/datasets/ytgangster/online-sales-in-usa
ss_df = pd.read_csv('./datasets/sales.csv')

display(ss_df)
```

### 데이터 전처리
- 사용하지 않을 feature 삭제
- 결측치 검사
- feature 명 변경
- 날짜 형식 변경
- 같은 의미인데 이름이 좀 다른 값 이름 통일
- RFM 구하기

```python
# 15개 카테고리
ss_df['category'].value_counts()
```

```python
# 사용하지 않을 feature 제거
ss_df = ss_df.drop(['sku', 'value', 'total', 'discount_amount', 'Discount_Percent', 'bi_st', 'ref_num'], axis=1)
```

```python
# 결측치 없음
ss_df.info()
```

```python
# feature 명 변경
ss_df.rename(columns={\
    'qty_ordered': 'quantity', 'payment_method': 'payment', 'Name Prefix':'name_prefix', 'First Name':'first_name'\
    , 'Middle Initial':'middle_initial', 'Last Name':'last_name', 'E Mail':'e_mail', 'Customer Since':'cust_since'\
    , 'Phone No.':'phone_num', 'Place Name':'place_name', 'User Name':'user_name'}, inplace=True)
ss_df.info()
```

```python
# 2021-09-30 ~ 2020-10-01
ss_df['order_date'].sort_values(ascending=False)
```

```python
# Recency 구하기
ss_df['date_point'] = pd.to_datetime(ss_df['order_date']).apply(lambda x: (pd.to_datetime('2021-09-30') - x).days)
ss_df
```

```python
# 같은 의미인데 이름만 다른게 많음
ss_df['status'].value_counts()
```

```python
# 구매 완료 여부를 같은 이름으로 묶어줌
ss_df.loc[ss_df['status'] == 'order_refunded', 'status'] = 'refund'

# 다른 방법으로 완료된것들을 complete로 변경
ss_df.loc[ss_df['status'] == 'received', 'status'] = 'complete'
ss_df.loc[ss_df['status'] == 'cod', 'status'] = 'complete'
ss_df.loc[ss_df['status'] == 'closed', 'status'] = 'complete'
ss_df.loc[ss_df['status'] == 'paid', 'status'] = 'complete'

# 대기중
ss_df.loc[ss_df['status'] == 'payment_review', 'status'] = 'pending'
ss_df.loc[ss_df['status'] == 'processing', 'status'] = 'pending'
ss_df.loc[ss_df['status'] == 'pending_paypal', 'status'] = 'pending'
ss_df.loc[ss_df['status'] == 'holded', 'status'] = 'pending'
```

```python
# 결제방식이 같은것 들 끼리 묶어줌
ss_df.loc[ss_df['payment'] == 'easypay_voucher', 'payment'] = 'Easypay'
ss_df.loc[ss_df['payment'] == 'Easypay_MA', 'payment'] = 'Easypay'
ss_df.loc[ss_df['payment'] == 'bankalfalah', 'payment'] = 'Easypay'
ss_df.loc[ss_df['payment'] == 'Payaxis', 'payment'] = 'Easypay'
ss_df.loc[ss_df['payment'] == 'jazzwallet', 'payment'] = 'Easypay'
ss_df.loc[ss_df['payment'] == 'jazzvoucher', 'payment'] = 'Easypay'
ss_df.loc[ss_df['payment'] == 'apg', 'payment'] = 'Easypay'
ss_df.loc[ss_df['payment'] == 'mcblite', 'payment'] = 'Easypay'
ss_df.loc[ss_df['payment'] == 'customercredit', 'payment'] = 'cod'
ss_df.loc[ss_df['payment'] == 'cashatdoorstep', 'payment'] = 'cod'
ss_df.loc[ss_df['payment'] == 'financesettlement', 'payment'] = 'cod'
```

```python
# age를 10단위로 분리
age_bins = [10, 20, 30, 40, 50, 60, 70, 80]
age_labels = ['10s', '20s', '30s', '40s', '50s', '60s', '70s']

ss_df['age_group'] = pd.cut(ss_df['age'], bins=age_bins, labels=age_labels, right=False)
# cs_sm_df = df.drop(columns=['age'], axis=1)
ss_df
```

```python
# Recency

# 마지막 결제일을 구해 Recency에 담아준다
ss_rfm_r = ss_df.groupby('user_name')['date_point'].min().reset_index(name='Recency')
ss_rfm_r
```

```python
# Frequency

ss_rfm_f = ss_df['user_name'].value_counts().reset_index(name='Frequency')
ss_rfm_f
```

```python
ss_rfm_f[ss_rfm_f['Frequency'] == 1]
```

```python
# 구매수량과 가격을 곱하여 total_price를 만듬
total_price = ss_df['quantity'] * ss_df['price']
display(total_price)

ss_df['total_price'] = total_price

ss_df
```

```python
# Monetary

# 각 유저를 기준으로 total_price를 더해준다.
ss_rfm_m = ss_df.groupby('user_name')['total_price'].sum().reset_index(name='Monetary')
ss_rfm_m
```

```python
# 테이블 병합
ss_rfm = pd.merge(ss_rfm_r, ss_rfm_f, on='user_name')
ss_rfm = pd.merge(ss_rfm, ss_rfm_m, on='user_name')
ss_rfm
```

```python
from sklearn.preprocessing import MinMaxScaler

#  MinMaxScaler()를 통해 0~1의 값으로 변환
scaler = MinMaxScaler()
ss_rfm[['Recency', 'Frequency', 'Monetary']] = \
                                    scaler.fit_transform(ss_rfm[['Recency', 'Frequency', 'Monetary']])
```

```python
ss_rfm
```

```python
# Recency를 1에서 빼서 최근에 구매한 사람이 가장 높은 점수를 받게 만듬
ss_rfm['Recency'] = 1 - ss_rfm['Recency']
ss_rfm
```

```python
# ss_rfm['Monetary'] = 2 * ss_rfm['Monetary']
# ss_rfm
```

```python
# RFM 점수 total
ss_rfm['TotalScore'] = ss_rfm[['Recency', 'Frequency', 'Monetary']].sum(axis=1)
ss_rfm
```

```python
# 기존 dataFrame과 sm_rfm을 병합 (cs_sm_df = shopping_mall기준으로 만든 테이블)
ss_rfm_df = pd.merge(ss_df, ss_rfm, on='user_name')
ss_rfm_df
```

```python
# Total 구한 후 필요 없는 RFM 제거
drop_columns = [
    'Recency', 'Frequency', 'Monetary'
]

ss_rfm_df = ss_rfm_df.drop(labels=drop_columns, axis=1)
ss_rfm_df
```

```python
import numpy as np

l1, l2, l3, l4 = np.percentile(ss_rfm_df['TotalScore'], [20, 40, 70, 90])
print(l1, l2, l3, l4)
```

```python
def get_level(x):
    if x <= l1:
        return 5
    if x <= l2:
        return 4
    if x <= l3:
        return 3
    if x <= l4:
        return 2
    return 1
```

```python
ss_rfm_df['Level'] = ss_rfm_df['TotalScore'].apply(get_level)
ss_rfm_df
```

```python
# 다섯개의 등급으로 분류
ss_rfm_df['Level'] = ss_rfm_df['Level'].replace([5, 4, 3, 2, 1], 
                                                ['Bronze', 'Silver', 'Gold', 'Diamond', 'VIP'])
```

```python
ss_rfm_df['Level'].value_counts()
```

```python
# 배송 완료
ss_df_complete = ss_rfm_df[ss_rfm_df['status'] == 'complete']
# 주문 취소
ss_df_canceled = ss_rfm_df[ss_rfm_df['status'] == 'canceled']
# 환불
ss_df_refund = ss_rfm_df[ss_rfm_df['status'] == 'refund']
# 주문취소 & 환불
ss_df_cr = ss_rfm_df[(ss_rfm_df['status'] == 'refund') & (ss_rfm_df['status'] == 'canceled')]
```

### 마케팅 전략 제안
- 가장 인기가 많은 Mobiles & Tablets가 브론즈 등급과 골드 등급이 상대적이게 많으며  
20대, 30대 고객과 50대 60대 고객이 가장 많기에 상대적 으로 가격에 예민한 20대와 30대를 겨냥하여 할인행사를 시행하고  
상대적으로 나이대가 좀 있는 50대와 60대를 겨냥하여 더 접근성을 높이기 위하여 상품의 설명을 쉽게 도와주거나 홈쇼핑 등 다양한 방법을 고려해볼 필요가 있어보임.  
- 50대의 Health & Sports 와 soghaat(선물)의 관심이 굉장히 많으며, 50대의 VIP가 가장 많고,  
카테고리 중 Health & Sports의 VIP가 가장 많다는 점을 고려하여 Health & Sports의 제품을 선물해주면 할인을 해주는 등  
Health & Sports와 soghaat를 엮어서 마케팅 하는 전략이 좋을 것 같다.  
- dekalb 지역의 VIP만 굉장히 많기때문에 해당 지역에서의 신규 회원 유입을 위한 홍보를 하고, 기존 VIP 고객 유치를 위해 VIP 등급의 혜택을 주는 전략이 좋을 것 같습니다.
- 취소와 환불의 전체 수치가 굉장히 높은데 취소 및 환불의 사유등을 조사하여 취소와 환불을 하지 않도록 만드는 전략을 찾아 볼 필요가 있을 것 같다.
